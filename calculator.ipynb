```python
from IPython.display import clear_output

def add(n1, n2):
    return n1 + n2


def subtract(n1, n2):
    return n1 - n2


def multiply(n1, n2):
    return n1 * n2


def divide(n1, n2):
    if n2 != 0:
        n3 =  n1 / n2
    else:
        n3 = 0
    return n3

def power(n1, n2):
    return n1 ** n2

operations = {"+": add, "-": subtract, "*": multiply, "/": divide, '^':power}


def calculator():
    correct_input = False
    while not correct_input:
        try:
            num1 = float(input("What's the first number?: "))
        except:
            print('Enter Nums ONLY!')
            continue
        else:
            for symbol in operations:
                print(symbol)
            correct_input = True
                
            continue_calculating = True
            while continue_calculating:
                correct_operation_symbol = False
                while not correct_operation_symbol:
                    operation_symbol = input("Pick an operation: ")
                    if operation_symbol in operations:
                        correct_operation_symbol = True

                while True:
                    try:
                        num2 = float(input("What's the next number?: "))
                    except:
                        print('Enter Nums ONLY')
                        continue
                    else:
                        calculation_function = operations[operation_symbol]
                        answer = calculation_function(num1, num2)
                        break

                print(f"{num1} {operation_symbol} {num2} = {answer}")
                end_calc = False
                while not end_calc:
                    conf = input(f"Type 'c' to continue calculating with {answer}, type 'n' to start a new calculation.\nor type 'e' to exit the calculator: ")
                    if conf == 'c':
                        num1 = answer
                        end_calc = True
                    elif conf == 'e':
                        continue_calculating = False
                        clear_output
                        end_calc = True
                    elif conf == 'n':
                        clear_output(wait=True) 
                        continue_calculating = False
                        calculator()
                        end_calc = True
                    else:
                        clear_output() 
                        print("Invalid Input!! \nInput a valid letter!")
                        
                            

calculator()
```
